## Imports

```python
import os
import xarray as xr
import numpy as np
```

```python
# Specify the converted netcdf file
netcdf_loc = os.path.join(
    os.path.expanduser('~'),
    'Downloads/02_geological_reference_model/UBC-GIF/NAC_density_reference.nc'
)
# load with xr and view
nc_data = xr.load_dataset(netcdf_loc)
# set the bad values to nan
nc_data = xr.where(nc_data==-100, np.nan, nc_data)
#view
nc_data
```

```python
# and plot
nc_data['data'].sel(Z=4500).plot()
```
